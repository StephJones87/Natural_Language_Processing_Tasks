```python
import pyLDAvis
import pyLDAvis.sklearn
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from collections import Iterable
from collections import Mapping
import pandas as pd
import re

pyLDAvis.enable_notebook()
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
df = pd.read_feather('rawvaccinetweetsfulltext.ftr', columns=None, use_threads=True);

print(df);
```

```python
raw_tweets = df.values.tolist()
```

```python
raw_tweets
```

```python
def remove_at_mentions(text):
  text = re.sub('([@#][\w_-]+)', '', text)
  return text

df['cleaned_tweet'] = df['tweet'].apply(lambda x: remove_at_mentions(x))
df.head(10)
```

```python
def https_removal(text):
  text = re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", text)
  return text

df['http_removed'] = df['cleaned_tweet'].apply(lambda x: https_removal(x))
df.head(10)
```

```python
def keepletters(text):
    text = re.sub('[^a-zA-Z ]+', '', text, re.UNICODE)
    return text

df['letters_only'] = df['http_removed'].apply(lambda x: keepletters(x))
df.head(10)
```

```python
df["lowercase_tweets"] = df["letters_only"].str.lower()
df.head(10)
```

```python
df_tweets = df['lowercase_tweets']
```

```python
df_tweets.head(10)
```

```python
docs_raw = df_tweets.values.tolist()
```

```python
tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                               stop_words = 'english',
                               lowercase = True,
                               token_pattern = r'\b[a-zA-Z]{3,}\b',
                               max_df = 0.5,
                               min_df = 10)

dtm_tf = tf_vectorizer.fit_transform(docs_raw)
```

```python
print(dtm_tf.shape)
```

```python
lda_tf = LatentDirichletAllocation(n_components=6, random_state=0)
```

```python
lda_tf.fit(dtm_tf)
```

```python
pyLDAvis.sklearn.prepare(lda_tf, dtm_tf, tf_vectorizer)
```
